1.5 categorical feature analysis

explore the possible relationship beween contract type and churn rates

```python
df.groupby('Contract').Churn.value_counts(normalize=True)
```

Contract Analysis:

- Month-to-month contracts have the highest churn rate
- Customers with longer contracts (one-year, two-year) have significantly lower churn rates

Insight:
There is a clear negative relationship between contract length and churn. Customers with longer-term commitments are less likely to churn.

 
Internet Service Analysis:

- Customers using Fiber optic have the highest churn rate
- DSL users have moderate churn
- Customers with no internet service have the lowest churn rate

Insight:

Customers with higher-tier internet services (especially Fiber optic) are more likely to churn compared to those with basic or no internet service.

`It may also indicate that Fiber optic services are priced higher or face stronger competition, leading to increased churn risk.`

```python
df.groupby('PaymentMethod').Churn.value_counts(normalize=True)
```

Payment Method Analysis:

- Customers using electronic check have the highest churn rate
- Other payment methods show relatively similar churn rates with no significant differences

Insight:

Payment method appears to have some impact on churn, particularly for electronic check users, while other methods do not show strong variation.

This may indicate that electronic check users are less engaged or less committed, or that this payment method is associated with a less stable customer segment.

It could also suggest that customers using automatic payment methods are more likely to stay due to convenience and lower friction.
