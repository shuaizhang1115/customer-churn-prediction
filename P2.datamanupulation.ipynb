1. Handle Missing Values

1.1 Checking hidden missing values

in the TotalCharges col, the data type is str, it is possible the blank data " " is not treated as NaN

```python
df['TotalCharges'].isin(["", " ", "NA", "null"]).sum()
```

We checked that there are 11 cells missing value. For TotalCharges, we need the data type to be numeric. By converting it, the missing value will be NaN

1.2 Convert hidden missng value

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.TotalCharges.info
```

1.3 Check missing values
now the col is changed to float and the missing value is NaN.

```python
df['TotalCharges'].isin(["", " ", "NA", "null"]).sum()
```

1.4  Fill missing values (median)
Median is robust to outliers

```python
df['TotalCharges'] =df.TotalCharges.fillna(df.TotalCharges.median())

df['TotalCharges']

df['SeniorCitizen'] = pd.to_numeric(df['SeniorCitizen'])
```

Step 2.2 Feature engineering


2.2.1 Looks at 'tenure' and 'TotalCharges', we can calculate avg charges every tenure year

```python
df['AvgCharges'] = df['TotalCharges'] / df['tenure'].replace(0,1)
```

2.2.2 We can further group the customer based on the tenure

```python
df['tenure_group']=pd.cut(df['tenure'], bins=[-1, 6, 12, 24, 36, 60, 72], labels=['0-6m', '6-12m', '1-2y','2-3y','3-5y', '5-6y'])
```

2.2.3 High Charge indicator

```python

df['high_charge'] = (df['MonthlyCharges'] > df['MonthlyCharges'].median()).astype(int)
```

2.2.4 Services



```python
services = [
    'PhoneService', 'MultipleLines', 
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]
```
